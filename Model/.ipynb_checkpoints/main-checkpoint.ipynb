# Importing the dependencies

```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
import imutils
import easyocr
import os
```

```python
data = cv2.imread(os.path.join("Image/car.jpg"))
```

```python
# data = cv2.resize(data, (500,500))
```

```python
plt.imshow(data)
plt.show()
```

```python
gray = cv2.cvtColor(data, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = "gray")
plt.show()
```

# Filter

```python
bfilter = cv2.bilateralFilter(gray, 11,17,17)
edged = cv2.Canny(bfilter, 30,200)
plt.imshow(edged, cmap = "grey")
```

```python
keypoints = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
countours = imutils.grab_contours(keypoints)
countours = sorted(countours, key = cv2.contourArea, reverse = True)[:10]
```

```python
location = None 
for countour in countours: 
    approx = cv2.approxPolyDP(countour, 10, True)
    if len(approx) == 4:
        location = approx
        break
location
```

```python
plt.imshow(cv2.rectangle(data, location[0][0], location[3][0], (0, 225, 0), 3))
```

```python
mask = np.zeros(gray.shape, np.uint8)
new_iamge = cv2.drawContours(mask, [location], 0,225,-1)
new_image = cv2.bitwise_and(data, data, mask = mask)
```

```python
plt.imshow(new_image, cmap = "grey")
```

```python
(x,y) = np.where(mask == 225)
(x1,y1) = (np.min(x), np.min(y))
(x2,y2) = (np.max(x), np.max(y))
cropped_image = gray[x1:x2+1, y1:y2+1]
plt.imshow(cropped_image, cmap = "grey")
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(cropped_image)
result
```
